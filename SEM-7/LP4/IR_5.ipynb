```python
import requests
from bs4 import BeautifulSoup
import yake
```

```python
def get_text(url):
  response = requests.get(url)
  soup = BeautifulSoup(response.content, 'html.parser')
  text_content = soup.find_all('p')
  text = ' '.join([para.get_text() for para in text_content])
  text = text.lower()
  return text
```

```python
def extract(url):
  text = get_text(url)
  kw_extractor = yake.KeywordExtractor(lan="en", n=1, top=100)
  keywords = kw_extractor.extract_keywords(text)
  for kw, score in keywords:
      array.append(kw)
```

```python
def display():
  print("Length: ", len(array))
  for kw in array:
    print(kw, ", ")
```

```python
extract('https://en.wikipedia.org/wiki/Education')
```

```python
extract('https://www.britannica.com/topic/education')
```

```python
extract('https://www.worldbank.org/en/topic/education')
```
